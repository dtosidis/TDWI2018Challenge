The dataset contains a list of destinations in any continent. The dataset is getting loaded into a graph database and afterwards natural language processing algorithms are used to recommend other matching destinations.

```python
from recommender.nlp import LanguageProcessing
from recommender.database import Database
from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors #the unsupervised algorithms within sklearn.neighbors
import gensim
import math
import numpy as np
import scipy as sp
import scipy.linalg as la
```

```python
d = Database()
L = LanguageProcessing(d)
words_text_mat = gensim.matutils.corpus2dense(L.corpus, 7396) # 159 docs, 7396 words
```

```python
# calculating the singular value decomposition (svd)
s = np.linalg.svd(words_text_mat, full_matrices=False, compute_uv=False)
#sorting them
s = -1*np.sort(-s)
# The squared sum of some of the singular values (eigenvalues) divided by the total sum is the variance 
# explained by those components 
s2 = s**2
cum_s = np.cumsum(s)/sum(s)
cum_s2 = np.cumsum(s2)/sum(s2)
print(cum_s2) # all components explain all 159 original dimensions
len(s)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(s[range(0,159,5)], range(0,159,5), '*-')
plt.xlabel('Threshold of singular values', fontsize=12)
plt.ylabel('Number of topics', fontsize=12)
plt.show()
```

This plot investigates  the  influence  of  the  decomposition level  on  the  search  results,  by  applying  the  search  at  the  best s, the  dimension  "number of topics",  in  the  SVD  algorithm  for  which  the best  results  are  obtained  and  threshold  values  at  different  levels of  decomposition. As we can see 40 out of 159 have more than zero thresholds and a decomposition level.

```python
# We want that the dimension reduction still explains a lot of the original data
plt.figure(figsize=(10,5))
plt.plot(range(0,159), cum_s2, '*-')
plt.xlabel('Number of topics', fontsize=12)
plt.ylabel('Variance explained', fontsize=12)
plt.show()
```

Checking the amount of variance in the data after computing the SVD can be used to determine the optimal number of topics to retain. The variance contained in the data can be viewed by plotting the singular values s in a scree plot. As we could see before 40 out of 159 topics have a threshold or more, but know we can observe the knee of the curve contains around 20 topics, which have variance lower than 0.9. We could apply it as the cut-off point for the number of topics to retain.

```python
# We model this tradeoff with an idea of minimum description length
# length of description =  (number of topics) / (number of texts)
# quality of description = variance explained = cumulative sum of squared singular values normed
# Idea best model is the one that explains the most with the smallest description length
# argmax quality of decscription - length of description
model_quality = -1*np.array(range(0,159))/159 + cum_s2
plt.figure(figsize=(10,5))
plt.plot(range(0,159), -1*np.array(range(0,159))/159 + cum_s2, '*-')
plt.xlabel('Number of topics', fontsize=12)
plt.ylabel('Model quality', fontsize=12)
plt.show()
```

We want to minimize the number of texts in order to find the quality of the topics as a whole. As we can see is a decreasing linear sequence after the highest value at the 20 topics. If we take the \"bell\" around the highest value, we could say that the 40 topics out of 159 have a good quality. After the distribution, we can see by using the argmax function the best quality is after 24 topics.

```python
#print(max(model_quality))
# Number of topics that this methodology recommends
print(np.argmax(model_quality))
```

```python
# idea of dimension reduction should still explain 70 percent of the variance (rule of thumb)
s07 = [i for i in cum_s2 if i <= 0.7]
print(s07)
print(len(s07))
```

Without the strict accuracy, we could assume the value of 0.7 is a low varianvce, which gives only 7 topics out of 159. So, the sample is very homogeneous and the similarity very high. For the recommender this makes the evaluation not very reliable, as we want the best destination.

```python
s = np.array([1.7320508, 1.4142135, 1.       ])
var = (s**2)/sum(s**2)
cum_var = np.cumsum(var)
print(cum_var)
print(-1 * np.array(range(1,len(s) + 1)) / len(s) + cum_var)
```

At first we print a variance of the array [1.7320508, 1.4142135, 1.       ] and we can see that it is from 0.5 to 1 the values. That means that the three values are associated from 50% to 100% and the sample has high homogeneity. The second prints a normalized variance my minimizing the length of them and the variance is from 0 to 16%. Consequently, the homogeneity is very low. This reveals a contradiction against the first sample That means, when we want better quality with smaller values, we are getting more accurate data.
